```python
# Imports
import pandas as pd
import numpy as np
```

```python
# Import DCAA CSV
dcaa = pd.read_csv('../data/dcaa.csv', header=None)
dcaa = dcaa.fillna('')
```

```python
dcaa
```

```python
# Reduce columns to FDA import

cols = [item for item in dcaa.iloc[[1]].values.tolist()[0]] # Get list of FDA headers

print(cols)
print('\n')

cols_index = sorted(list(set([cols.index(item) for item in cols if item != '']))) # Get index of headers and remove unnecessary columns; must be a better way to do this
print(cols_index)

fda = dcaa[cols_index]

fda
```

```python
# Remove non-FDA headers

fda.columns = fda.iloc[1]
fda = fda[3:]
fda = fda.reset_index(drop=True)
fda
```

```python
# Select only rows that have not been uploaded
# These are files for which there is:
# 1. No FDA handle in dc.identifier.uri
# 2. A rights statement

fda_upload = fda[fda['dc.identifier.uri'] == '']
fda_upload = fda_upload[fda['dc.rights'] != '']
fda_upload
```

```python
import os 

def add_thumbnail_filename(pdf_filename):
    """
    Convert pdf filename like 'isawdca_000805.pdf'
    to cell contents for FDA upload like 'isawdca_000805.pdf||isawdca_000805_thumb.png'
    """
    filename, file_extension = os.path.splitext(pdf_filename)
    contents = f'{pdf_filename}||{filename}_thumb.png'
    return contents


fda_upload['filename'] = fda_upload['filename'].apply(lambda x: add_thumbnail_filename(x))
```

```python
# Remove identifier column

fda_upload = fda_upload.drop('dc.identifier.uri', axis=1)
```

```python
# Sort dataframe
fda_upload = fda_upload.sort_values(['filename'])
```

```python
# Export dataframe to csv

fda_upload.to_csv('../data/out/fda_upload.csv', index=False)
```
