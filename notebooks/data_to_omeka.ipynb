```python
# Imports

from pprint import pprint

import pandas as pd
import numpy as np
```

```python
# Import DCAA CSV

dcaa = pd.read_csv('../data/dcaa.csv', header=None)
dcaa = dcaa.fillna('')
```

```python
dcaa
```

```python
# Reduce columns to Omeka import

cols = [item for item in dcaa.iloc[[2]].values.tolist()[0]] # Get list of FDA headers\
cols_index = [i for i, item in enumerate(cols) if item != '']
omeka = dcaa[cols_index]

# Remove non-Omeka headers
omeka.columns = omeka.iloc[2]
omeka = omeka[3:]
omeka = omeka.reset_index(drop=True)

# Select only rows that have not been uploaded
omeka = omeka[(omeka.iloc[:, 0] != '') & (omeka.iloc[:, 1] == '')]

omeka
```

```python
import re

def concatenate_columns(col_list):
    temp = zip(*col_list)
    temp = list(temp)
    temp = [tuple(filter(None, tp)) for tp in temp]
    temp = [" | ".join(item) for item in temp]
    temp = [re.sub(r'\s+\|\s+', '|', item) for item in temp]
    return temp
```

```python
# Helper functions for applys

def makeHTML(url):
    if url:
        return "<a href='" + url + "'>" + url + "</a>"

def makeOCLC(id):
    if id:
        return "<a href='http://www.worldcat.org/oclc/" + id.strip() + "'>OCLC: " + id + "</a>"

def makeRelationFull(string):
    if string:
        temp = string
        temp = temp.replace('http://hdl.handle.net/','full:001:https://archive.nyu.edu/bitstream/')
        temp = temp.replace('|','/1/')
        return temp    
    
def makeRelationThumb(string):
    if string:
        temp = string
        temp = temp.replace('http://hdl.handle.net/','thumb:001:https://archive.nyu.edu/bitstream/')
        temp = temp.replace('|','/2/')
        temp = temp.replace('.pdf','_thumb.png')
        return temp
```

```python
# Column applys

omeka.iloc[:, 0] = omeka.iloc[:, 0].apply(lambda x: makeHTML(x))
omeka.iloc[:, 3] = omeka.iloc[:, 3].apply(lambda x: makeOCLC(x))

omeka['Dublin Core:Relation1'] = omeka.iloc[:, 0] + '|' + omeka['Dublin Core:Relation']
omeka['Dublin Core:Relation2'] = omeka.iloc[:, 0] + '|' + omeka['Dublin Core:Relation']
omeka['Dublin Core:Relation1'] = omeka['Dublin Core:Relation1'].apply(lambda x: makeRelationFull(x))
omeka['Dublin Core:Relation2'] = omeka['Dublin Core:Relation2'].apply(lambda x: makeRelationThumb(x))
```
